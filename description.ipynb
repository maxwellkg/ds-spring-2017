```python
#Import data for description column
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn import metrics
import pickle

import mord
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_pickle('/Users/Oliver/Downloads/pickled_data.p')
data.head()
```

```python
x = [col for col in data.columns if col not in ['bathrooms', 'latitude', 'longitude', 'building_id', 'created', 'street_address', 'description', 'display_address', 'features', 'listing_id', 'manager_id', 'photos', 'interest_level', 'neighborhood']]
X = data[x]
Y = data['interest_level'].replace(['low', 'medium', 'high'], [1,2,3])
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=.75)
```

```python
#Random Forest
random_forest = RandomForestClassifier()
random_forest = random_forest.fit(X_train, Y_train)

print("Random Forest MAE = %.3f" % metrics.mean_absolute_error(random_forest.predict(X_test), Y_test))

#LogisticIT
mord_it = mord.LogisticIT()
mord_it = mord_it.fit(X_train, Y_train)

print("Logistic IT MAE = %.3f" % metrics.mean_absolute_error(mord_it.predict(X_test), Y_test))

#Logistic Regression
logistic = LogisticRegression()
logistic.fit(X_train, Y_train)

print("Logistic MAE = %.3f" % metrics.mean_absolute_error(logistic.predict(X_test), Y_test))

#GaussianNB
nb = GaussianNB()
nb.fit(X_train, Y_train)

print("GaussianNB MAE = %.3f" % metrics.mean_absolute_error(nb.predict(X_test), Y_test))
```
